```r
# Load necessary libraries
library(ggplot2)
```

```r
# Read the data
crime_data <- read.csv("/kaggle/input/crime-data/crimedata.csv")
```

```r
# Bar Chart - Number of Burglaries by State
ggplot(crime_data, aes(x = state, y = burglaries)) +
  geom_bar(stat = "identity") +
  theme(axis.text.x = element_text(angle = 90, hjust = 1)) +
  labs(title = "Burglaries by State", x = "State", y = "Number of Burglaries")
```

```r
# Pie Chart - Population by State
state_population <- aggregate(population ~ state, data = crime_data, sum)

pie(state_population$population, labels = state_population$state, 
    main = "Population Distribution by State")
```

```r
# Histogram - Population distribution
ggplot(crime_data, aes(x = population)) +
  geom_histogram(binwidth = 10000, fill = "blue", color = "black") +
  labs(title = "Population Distribution", x = "Population", y = "Frequency")
```

```r
# Time Line Chart - Population over Community Code (as a proxy for time)
ggplot(crime_data, aes(x = communityCode, y = population)) +
  geom_line() +
  labs(title = "Population Trend over Community Codes", x = "Community Code", y = "Population")
```

```r
# Scatter Plot - Population vs Violent Crimes
ggplot(crime_data, aes(x = population, y = ViolentCrimesPerPop)) +
  geom_point(color = "blue") +
  labs(title = "Population vs Violent Crimes per Pop", x = "Population", y = "Violent Crimes Per Pop")
```

```r
# Bubble Plot - Population vs Violent Crimes and Non-Violent Crimes
ggplot(crime_data, aes(x = ViolentCrimesPerPop, y = nonViolPerPop, size = population)) +
  geom_point(alpha = 0.5, color = "red") +
  labs(title = "Violent Crimes vs Non-Violent Crimes", x = "Violent Crimes Per Pop", y = "Non-Violent Crimes Per Pop", size = "Population")
```
